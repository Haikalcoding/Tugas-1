```python
#queue
nama = []
max_size = int(input("masukkan max nama :"))

def isFull ():
  if len(nama)== max_size :
    print ("buah penuh")
    return True
  return False

def enqueue():
  if isFull() == False :
    i = input("masukkan nama baru:")
    nama.append (i)
    print (i,"dimasukkan")

def isEmpty():
  if len(nama)==0:
    print("nama kosong")
    return True
  return False

def dequeue():
  if isEmpty ()== False:
    print(nama.pop(0))

perintah = 0
while perintah != 5:
  print("""1. enqueue -> tambah
  2. dequeue -> hapus 
  3. full -> cek nama apakah penuh
  4. size -> melihat ukuran
  5. stop""")

  perintah = int(input("masukkan perintah :"))

  if perintah == 1:
    enqueue ()
  elif perintah == 2:
    dequeue ()
  elif perintah == 3:
    if isFull()==False:
      print("nama belum penuh")
  elif perintah == 4:
    print ("isi nama saat ini:",len(nama),"\n  maksimal nama:",max_size)
```
